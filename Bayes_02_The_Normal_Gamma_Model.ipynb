### 3. What distribution do you think this $\theta$ has?

$\large
\begin{align*}
\prod_{i=1}^n\frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{1}{2\sigma^2}(x_i-\theta)^2} &={} (2\pi\sigma^2)^{-\frac{n}{2}} e^{-\frac{1}{2\sigma^2} \sum_{i=1}^n (x_i-\theta)^2}\\
&\propto{} e^{-\frac{1}{2\sigma^2} \sum_{i=1}^n \theta^2 -2x_i\theta }\\

\textrm{So} & \;\;\; \theta \sim N\left(\bar x, \frac{\sigma^2}{n}]\right)
\end{align*}
$

```python
from scipy import stats
n = 1000  # num of observations
samples = stats.norm(loc=0, scale=1).rvs(size=n)

posterior = stats.norm(loc=samples.mean(), scale=1/(n)**0.5)

# To do Bayesian inference, sample from the posterior and then calculate a credible interval
# Bayesian inference is an interval estimate. You have to provide a credible interval to get marks
m = 10000
theta = posterior.rvs(size=m)
theta

import matplotlib.pyplot as plt
import numpy as np

theta_support = np.linspace(-0.2, 0.2, 1000)
plt.hist(theta, density=True)
plt.plot(theta_support, posterior.pdf(theta_support))
plt.show()
```

## Part C (II): Full Conditional [Posterior] Distributions

### 1. What is $p(\theta|\tau, \mathbf{x})$?<br>This is a so-called a *full conditional* [posterior distribution]

#### *This is actually just asking for $p(\theta|\tau, \mathbf{x}) \propto p(\textbf{x}, \theta|\tau) \propto p(\textbf{x},\theta,\tau)$*<br>which is just $p(\textbf{x},\theta,\tau)$ viewed only as a function of $\theta$ with $\textbf{x}$ given<br>*(which means $\textbf{x}$ and $\tau$ are known because they've been observed)*

### 2. What is $p(\theta,\tau | \mathbf{x})$ this if all normalizing constants are removed if ONLY $\tau$ is was a random variable?

#### *This is actually just asking for $p(\tau | \theta, \mathbf{x}) \propto p(\textbf{x},\tau |\theta) \propto p(x,\theta,\tau)$*<br>which is just $p(\textbf{x},\theta,\tau)$ viewed only as a function of $\tau$ with $\textbf{x}$ given<br>*(which means $\textbf{x}$ and $\theta$ are known because they've been observed)*

### That is, what is the *full conditional* [posterior distribution] $p(\tau|\theta, \mathbf{x})$?

$\Large
\begin{align*}
p(\tau|\theta, \mathbf{x}) &\propto{} f(\mathbf{x}|\theta)p(\tau|\theta)=f(\mathbf{x}|\theta)p(\tau)\\
&\propto{} \prod_{i=1}^n\sqrt{\frac{\tau}{2\pi}} e^{-\frac{\tau}{2}(x_i-\theta)^2} \times 1 \\
&\propto{} \tau^{\frac{n}{2}} e^{-\frac{\tau}{2}\sum_{i=1}^n(x_i-\theta)^2}
\end{align*}
$


$\Large \tau \sim \large \textrm{gamma}\bigg(\alpha = \frac{n}{2}+1, \textrm{SCALE}= \left(\frac{1}{2} \sum_{i=1}^n (x_i-\mu)^2\right)^{-1} \bigg)$

### 3. What is $p(\tau|\theta, \mathbf{x})$ if we still have $p(\tau) \propto 1$<br>but now $p(\theta)$ is $N(\theta_0, \tau^2_0)$?

## Part C (III): Gibbs Sampling

The canonical example of ***Gibbs sampling*** is the ***normal-gamma*** model and here's a simplified version with $p(\theta,\tau) \propto 1$

$\begin{align*}
p\left(x|\theta,\tau=\frac{1}{\sigma^2}\right) = {} & \prod_{i=1}^n \sqrt{\frac{\tau}{2\pi}} e^{-\frac{\tau}{2}(x_i-\theta)^2}\\
p(\theta,\tau) \propto {} & 1 \\
p(\theta,\tau| x) \propto p(\theta,\tau, x) \propto {} &  \tau^{n/2} e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i-\theta)^2} \\
p(\tau|\theta, x) = {} & \text{Gamma}\left(\alpha = \frac{n}{2}+1, \textrm{rate} = \frac{1}{2} \sum_{i=1}^n (x_i-\theta)^2 \right)\\
p(\theta| \tau, x)  \propto {} & e^{-\frac{\tau}{2} \sum_{i=1}^n \theta^2 - 2\theta x_i} =  e^{-\frac{\tau}{2} \left(n\theta^2 - 2\theta \sum_{i=1}^n x_i\right)} = e^{-\frac{\tau n }{2} \left(\theta^2 - 2\theta \bar x\right)}\\
  \propto {} & e^{-\frac{n \tau}{2} (\theta - \bar x)^2} \\
p(\theta| \sigma^2 = \tau^{-1}, x) = {} & N\left(E[\theta] = \bar x, Var(\theta) = \frac{\sigma^2}{n}\right)
\end{align*}$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# loc is mean and scale is standard deviation
n = 30; x = stats.norm(loc=0, scale=1).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples
theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization
for c in range(C):
    for g in range(1,G):
        # https://en.wikipedia.org/wiki/Gamma_distribution
        tau[c,g] = stats.gamma(a=n/2+1, scale=2/((x - theta[c, g-1])**2).sum()).rvs() # \NOTE: DONE IN CLASS
        theta[c,g] = stats.norm(x.mean(), scale=1 / (tau[c,g] * n)**0.5).rvs() #\ NOTE: DONE IN CLASS
```

```python
fig,ax = plt.subplots(1, 4, figsize=(14,3)); ax[0].set_title("$\\theta$ Burn-In"); ax[1].set_title("$\\tau$ Burn-In"); ax[2].set_title("$\\theta$ Converged Markov Chain\nStationary Distribution Samples"); ax[3].set_title("$\\tau$ Converged Markov Chain\nStationary Distribution Samples")
burn = 20
demo = 120
for c in range(C):
    ax[0].plot(theta[c,:burn], label="$\\theta$ Chain "+str(c))
    ax[1].plot(tau[c,:burn], label="$\\tau$ Chain "+str(c))
    ax[2].plot(np.arange(burn, demo, dtype=int), theta[c,burn:demo], label="$\\theta$ Chain "+str(c))
    ax[3].plot(np.arange(burn, demo, dtype=int), tau[c,burn:demo], label="$\\tau$ Chain "+str(c))
ax[0].legend(); ax[1].legend(); ax[2].legend(); ax[3].legend();
```

```python
# loc is mean and scale is standard deviation
m = 80; s = 10
n = 300; x = stats.norm(loc=m, scale=s).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples
theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization
for c in range(C):
    for g in range(1,G):
        # https://en.wikipedia.org/wiki/Gamma_distribution
        shape_tau = n/2 + 1
        scale_tau = 2 / ((x - theta[c, g-1])**2).sum()
        tau[c,g] = stats.gamma(a=shape_tau, scale=scale_tau).rvs()

        mean_theta = x.mean()
        sd_theta = 1 / np.sqrt(n * tau[c, g])
        theta[c,g] = stats.norm(loc=mean_theta, scale=sd_theta).rvs()
```

### Q2

Derive the *full conditional distributions* $p(\tau|\theta, \mathbf{x})$ and $p(\theta|\tau, \mathbf{x})$ if $p(\theta)$ is $N(\theta_0, \tau_0)$ and $p(\tau) \propto \tau^{\frac{\alpha}{2}-1} e^{- \frac{\lambda}{2} \tau}$.

*Explain how to equate $\alpha$ and $\lambda$ with controlling the amount and nature of information the prior contributes to the posterior. How can we interpret these as controlling a "number of data points" contributed by the prior relative to the number of data points contributed to the posterior through the likelihood?*

Create an updated version of the *Gibbs sampler* for $N(\theta_0, \tau_0)$ and $\textrm{gamma}(\textrm{shape}=\frac{\alpha}{2}, \textrm{rate}=\frac{\lambda}{2})$ priors for $p(\theta)$ and $p(\tau)$ respectively.

*Demonstrate how changing the prior hyperparameters influences the posterior relative to the sample and the likelihood.*


**Q2 Answer**

### **Full Conditional $p(\tau|\theta, \mathbf{x})$ for Precision $\tau$**
Prior $p(\tau) \propto \tau^{\frac{\alpha}{2}-1} e^{- \frac{\lambda}{2} \tau}$

Likelihood $p(\mathbf{x}|\theta,\tau) \propto \tau^{n/2} e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i - \theta)^2}$

Using the likelihood and prior:
\begin{align}
p(\tau|\theta,\mathbf{x}) &\propto \tau^{n/2} e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i-\theta)^2} \times \tau^{\frac{\alpha}{2}-1} e^{-\frac{\lambda}{2} \tau}\\
&=\tau^{\frac{\alpha + n}{2} - 1} e^{-\frac{\tau}{2} \left( \lambda + \sum_{i=1}^n (x_i - \theta)^2 \right)}\\
\end{align}
Therefore, $p(\tau|\theta,\mathbf{x}) = \text{Gamma}\left( \frac{\alpha + n}{2}, \frac{\lambda + \sum_{i=1}^n (x_i - \theta)^2}{2} \right)$.

---

### **Full Conditional for $\theta$**
Prior $\theta \sim N(\theta_0, \tau_0) \propto \exp\left(-\frac{1}{2\tau_0}(\theta - \theta_0)^2\right)
$

Likelihood $p(\mathbf{x}|\theta,\tau) \propto \exp\left(-\frac{\tau}{2} \sum_{i=1}^n (x_i - \theta)^2\right)$

Using the likelihood and prior:

$p(\theta|\tau,\mathbf{x}) \propto \exp\left(-\frac{\tau}{2} \sum_{i=1}^n (x_i-\theta)^2\right) \cdot \exp\left(-\frac{1}{2\tau_0}(\theta - \theta_0)^2\right)$

Expanding the left likelihood term:
$\sum_{i=1}^n (x_i - \theta)^2 = n\theta^2 - 2\theta n\bar{x} + \text{constant terms not involving } \theta$

Expanding the prior term: $(\theta - \theta_0)^2 = \theta^2 - 2\theta\theta_0 + \theta_0^2$.
Note that we can omit the constant last term that does not involve $\theta$.

Combining the two:
\begin{align}
p(\theta|\tau,\mathbf{x}) &\propto \exp\left(-\frac{\tau}{2} \sum_{i=1}^n (x_i-\theta)^2\right) \cdot \exp\left(-\frac{1}{2\tau_0}(\theta - \theta_0)^2\right)\\
&\propto \exp\left(-\frac{\tau}{2} \left(n\theta^2 - 2\theta n\bar{x}\right) - \frac{1}{2\tau_0}\left(\theta^2 - 2\theta\theta_0\right)\right)\\
&= \exp\left(-\frac{1}{2}\left[ \tau n\theta^2 - 2\tau n\bar{x}\theta + \frac{1}{\tau_0}\theta^2 - \frac{2}{\tau_0}\theta\theta_0 \right]\right)\\
&= \exp\left(-\frac{1}{2}\left[ \left(\tau n + \frac{1}{\tau_0}\right)\theta^2 - 2\theta\left(\tau n\bar{x} + \frac{\theta_0}{\tau_0}\right) \right]\right)
\end{align}

Precision = $\tau n\bar{x} + \frac{\theta_0}{\tau_0}$

Mean = $\frac{\tau n\bar{x} + \frac{\theta_0}{\tau_0}}{\tau n + \frac{1}{\tau_0}}$

This is the kernel of a normal distribution.

$p(\theta|\tau,\mathbf{x}) = N\left( \frac{\tau n\bar{x} + \frac{\theta_0}{\tau_0}}{\tau n + \frac{1}{\tau_0}}, \; \frac{1}{\tau n + \frac{1}{\tau_0}} \right)$

---

### **Explanation for $\alpha$ and $\lambda$**

$\alpha$ controls the weight of prior evidence, equivalent to having a larger or smaller number of observations contributed by the prior. $\lambda$ controls the variance (i.e. the spread/scale) of the prior evidence. Larger $\alpha$ and $\lambda$ give the prior more influence on the posterior, whereas smaller values give the data points more of a influence/contribution to the posterior since the data can dominate more easily.

---

### ***Code and Demonstration below.***
I select a weak set of prior hyperparameters and then a strong set of prior hyperparameters.

As you can see, the different sets of prior hyperparameters can substantially change how much a similar amount of observed data from the same distribution can affect the posterior distribution.

Smaller alpha and lambda, larger tau and theta closer to the true mean allows the data to dominate the prior and the sampled data strongly shapes the posterior.

Conversely, larger alpha and lambda, smaller tau and theta further from the true mean allows the prior to strongly shape the posterior.

### 1. For a proper prior, the posterior is proportional to the joint

**Bayes’ Theorem** states
  $p(\theta \mid \mathbf{x})\;=\;\frac{p(\mathbf{x} \mid \theta)\,p(\theta)}{p(\mathbf{x})}.$

When the prior $p(\theta)$ is a proper distribution (i.e. integrates to 1), and the data $\mathbf{x}$ is given, the term $p(\mathbf{x})$ is *just a constant* (the marginal likelihood).  Hence,

$p(\theta \mid \mathbf{x}) \propto p(\mathbf{x} \mid \theta)\;p(\theta)$.

What this means is that for a proper prior, the posterior is simply proportional to the product of the prior and the likelihood. The denominator $p(\mathbf{x}) = \int p(\mathbf{x}|\theta)\,p(\theta)\,d\theta$ is precisely the normalization constant that ensures the posterior integrates to 1.

### 2. Describing how to ientify the exact posterior by inspecting the joint

Because
$p(\theta\mid \mathbf{x}) \propto p(\mathbf{x}\mid \theta)\,p(\theta)$,
we can inspect the form of this product and see if it matches a known distribution family. For example, if $p(\theta)$ is $\mathrm{Beta}(a,b)$ and $p(\mathbf{x}\mid \theta)$ is Binomial (or Bernoulli i.i.d. form), their product matches a Beta distribution again.  

Thus, by looking at the joint $p(\mathbf{x},\theta) = p(\mathbf{x}\mid\theta)\,p(\theta)$ and identifying the functional form in $\theta$, you can pinpoint the distribution family and the correct normalizing constant without doing the full integral. This is the idea of conjugacy: the posterior has the same distributional form as the prior, just with updated parameters.

More generally, even when it’s not a conjugate prior, the same principle holds: the posterior is that product, normalized. Sometimes we can’t write a neat closed‐form, but the proportional relationship still holds.


## 3. Why this is beside the point in practice

Bayesian inference in modern practice rarely relies on an explicit formula. Instead, we:

1. Draw samples from the posterior distribution (e.g. via Gibbs sampling, Metropolis-Hastings, etc.).  
2. Make histograms of those samples with credible intervals of probabilities.  
3. The credible intervals tell us the probability that $\theta$ is in [$a, b$] is 95%.”

Hence, even if the exact functional form is complicated, we can *numerically approximate* it by sampling. The posterior’s normalizing constant isn’t something we need to solve in closed form.

While it is conceptually correct and useful in simple cases to identify the distribution by inspection of the joint, in general we simply rely on drawing from the posterior. This avoids performing messy integrals or normalizing constants by hand, fulfilling the practical needs of Bayesian inference.
